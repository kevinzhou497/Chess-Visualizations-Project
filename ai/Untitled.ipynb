```python
from chess_ai import *
player = get_chess_player()
```

```python
root_fen = ChessEnv().reset().board.fen()

# build tree of depth 6, branching factor 3 from starting position
n = ChessTreeNode(root_fen, player).build_tree(
    root_fen,
    B=3,
    depth=6
)
```

```python
# tree (incomplete) of depth 6, branching factor 3
print(n.pprint())
```

```python
# most likely state after 6 moves
n.most_likely_state(6)
```

```python
# probability map of occupying each cell (R: white, G: black)
n.heatmap_for_states(n.bfs_depth_D(6))
```

### Using stockfish

```python
from stockfish import Stockfish

stockfish = Stockfish("../stockfish_13_linux_x64")
fen1 = '4r3/1b2n1pk/1p1r1p2/pP1p1Pn1/P1pP1NPp/2N1P2P/2R3B1/5RK1 b - - 0 1' # alphazero game
fen2 = '4r1k1/2n1bppp/p3p3/1p6/8/1P2B1P1/P3PPBP/R5K1 w - - 1 1'
```

```python
# search with single best move
node = StockfishTreeNode(fen1, stockfish, hack_probs=False).build_tree(fen1, depth=8)
print(node.pprint())
```

```python
# hacky way of getting probabilities and branching
node = StockfishTreeNode(fen1, stockfish, hack_probs=True).build_tree(fen1, B=3, depth=5)
print(node.pprint())
```

```python
# look at moves
node = StockfishTreeNode(fen1, stockfish, hack_probs=False).build_tree(fen1, depth=8)
for i in range(1, 9):
    node.most_likely_state(i)
```
